```python
#Dijkstra Algorithm
n = 6
a = [[0,2,5,1,float('inf'),float('inf')],
     [2,0,3,2,float('inf'),float('inf')],
    [5,3,0,3,1,5],
     [1,2,3,0,1,float('inf')],
     [float('inf'),float('inf'),1,1,0,2],
     [float('inf'),float('inf'),5,float('inf'),2,0]
    ]

v = [False] * n
d = [0 for _ in range(n)]

def getSmallIndex():
    minimum = float('inf')
    index = 0
    for i in range(n):
        if(d[i] < minimum and not v[i]):
            minimum = d[i]
            index = i
    return index

def dijkstra(start):
    #초반 d List 값 초기화
    for i in range(n):
        d[i] = a[start][i]
    v[start] = True
    
    for j in range(n-2):
        current = getSmallIndex()
        v[current] =True
        for k in range(n):
            if(not v[k]):
                if(d[current] + a[current][k] <d[k]):
                    d[k] = d[current] + a[current][k]

#run
dijkstra(0)
for i in d:
    print(i,end=" ")
    
```
