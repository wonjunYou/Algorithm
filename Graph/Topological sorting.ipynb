```python
from collections import deque

#N명의 학생,  M회 비교 횟수.
#A B의 비교 결과
N, M = map(int, input().split())

edge_count =[ 0 for i in range(32001)]
graph = [[] for i in range(32001)]
arr= list()
queue = deque()

for i in range(M):
    a, b = map(int, input().split())
    arr.append([a, b])

for a,b in arr:
    edge_count[b] +=1
    graph[a].append(b)
    
#Make first queue
for i in range(1,N+1):
    if(edge_count[i] == 0):
        queue.append(i)
        
while queue:
    popped = queue.popleft()
    for j in graph[popped]:
        edge_count[j] -=1
        if edge_count[j] == 0:
            queue.append(j)
    print(popped,end=" ")
```
