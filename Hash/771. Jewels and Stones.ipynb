```python
'''
https://leetcode.com/problems/jewels-and-stones/

771. Jewels and Stones

Runtime: 32 ms, faster than 54.79% of Python3 online submissions for Jewels and Stones.
Memory Usage: 14.2 MB, less than 44.97% of Python3 online submissions for Jewels and Stones.

'''

class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        table = {}
        count = 0
        
        #update stone's character frequency info into the table.
        for char in stones:
            if char in table:
                table[char] +=1
                
            else:
                table[char] = 1
            
        #sum all counts in jewels's char by using table.
        for char in jewels:
            if char in table:
                count += table[char]
        
        return count

```

```python
'''
Sol #2) use defaultdict()

https://leetcode.com/problems/jewels-and-stones/

771. Jewels and Stones

Runtime: 32 ms, faster than 54.79% of Python3 online submissions for Jewels and Stones.
Memory Usage: 14.4 MB, less than 11.71% of Python3 online submissions for Jewels and Stones.

'''

class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        table = collections.defaultdict(int)
        count = 0
        
        #useless if condition because of using defaultdict
        for char in stones:
            table[char] +=1
            
        for char in jewels:
            count += table[char]
        
        return count
            
```

```python
'''
Sol #3) use Counter()

https://leetcode.com/problems/jewels-and-stones/

771. Jewels and Stones

Runtime: 24 ms, faster than 95.11% of Python3 online submissions for Jewels and Stones.
Memory Usage: 14.2 MB, less than 73.57% of Python3 online submissions for Jewels and Stones.

'''

class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        table = collections.Counter(stones)
        count = 0
    
        for char in jewels:
            count += table[char]
        
        return count
```

```python
'''
Sol #4) use list comprehension

https://leetcode.com/problems/jewels-and-stones/

771. Jewels and Stones

Runtime: 32 ms, faster than 54.79% of Python3 online submissions for Jewels and Stones.
Memory Usage: 14.4 MB, less than 11.71% of Python3 online submissions for Jewels and Stones.

'''

class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        return len([char for char in stones if char in jewels])
```
