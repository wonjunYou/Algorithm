```python
#https://leetcode.com/problems/valid-parentheses/

#20. Valid Parentheses

#Sol 1. using stack solution

#Runtime: 56 ms
#Memory Usage: 16.4 MB

class Solution:
    def isValid(self, s: str) -> bool:
        stack = list()
        match_table = {
            '}':'{',
            ')':'(',
            ']':'['
        }

        #match가 되는지 체크, match가 안된다면 stack에 추가
        for i in s:
            if stack and i in match_table:
                if stack.pop() != match_table[i]:
                    return False
                continue
            stack.append(i)
  
        if stack:
            return False
        return True
                    
                
        
        
```
