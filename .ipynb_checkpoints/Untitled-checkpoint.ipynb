```python
#행렬 경로 문제

def matrixPath(n):
    for i in range(n):
        M[i][0] = 0;
        
    for j in range(n):
        M[0][j] = 0;
    
    for i in range(n):
        for j in range(n):
            M[i][j] = M[i][j] + max(M[i-1][j] , M[i][j-1])
    
    return M[i][j]
    
n= int(input("숫자를 입력하세요: "))
M=[[0]*n for i in range(n)]
print(matrixPath(n))

```
