```python
#Fractional Knapsack Problems

KWeight = 5
value = [12,10,20,15]
weight= [2,1,3,2]
count = 0
value_per_weight = [0]*100
maxValue = 0

for i in range(len(value)):
    for j in range(weight[i]):
        value_per_weight[j+count] = value[i]/weight[i]
        count += 1

value_per_weight.sort(reverse=True)
for k in range(KWeight):
    maxValue += value_per_weight[k]
print(maxValue)
```
