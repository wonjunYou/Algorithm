```python
p = [20,10,5,10]
f= [3,0,0,0]

score = [0]*(len(p)+1)
n = len(p)
jump = 0

for i in range(n-1, -1 ,-1):
    tmp = p[i]
    
    if(f[i]+i+1 < n):
        tmp = tmp+score[f[i]+i+1]
        jump = score[i+1]
    
    score[i]= max(tmp,jump)

print("누적 최고 점수는",score[0])
        
```
