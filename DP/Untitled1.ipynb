```python
#문제 점프
p = [20,10,5,10]
f= [3,0,0,0]

score = [0]*(len(p)+1)
n = len(p)

for i in range(n-1,-1,-1):
    tmp = p[i]
    
    if(f[i]+i+1 < n):
        tmp = tmp + score[f[i] + i + 1]
    #jump면 어차피 안 풀었을 것.
    jump = score[i+1]
    
    score[i] = max(tmp,jump)
print(score[0])
```

```python
#LCS
a = "ABCDGH"
b = "AEDFHR"

m = len(a)
n = len(b)

cache = [[0] *(m+1) for _ in range(n+1)]

for i in range(m+1):
    for j in range(n+1):
        if((i==0) and (j==0)):
            cache[i][j] = 0
        elif(a[i-1] == b[j-1]):
            cache[i][j] = cache[i-1][j-1] + 1
        else:
            cache[i][j] = max(cache[i-1][j], cache[i][j-1])
print(cache[m][n])
```

```python
#LIS
N = int(input())
sequence = list(map(int, input().split()))

# dp[i] : i번째에서 가장 긴 부분 수열 
dp = [0] * N

# 모든 수를 따짐
for i in range(0, N):
    min_value = 0
    for j in range(i):
        # 현재 수가 더 큰 경우는 작은 수의 경우의 수에다가 + 1을 해준다.
        if(sequence[i] > sequence[j]):
            min_value = max(min_value, dp[j])
    dp[i] = min_value + 1
print(max(dp))
```

```python
#Representative 3 problems in Dynamic programming,
#1. 행렬경로

Matrix = [[2,3,8,4],[9,1,11,2],[6,5,14,12],[9,3,5,1]]
n = int(input())

cache = [[0]*(n+1) for _ in range(n+1)]

for i in range(1,n+1):
    for j in range(1,n+1):
        cache[i][j] = Matrix[i][j] + max(cache[i-1][j], cache[i][j-1])
print(cache[n][n])
```

```python
#2.행렬의 곱셈

Matrix = [40,20,30,15,5,35]
N=5
cache =[[0]*(N+1) for _ in range(N+1)]

for i in range (2,N+1):
    for j in range(1,N-i+2):
        k = j+i-1
        min_value=0
        for r in range(j,k):
            result = cache[j][r]+cache[r+1][k]+Matrix[j]*Matrix[r+1]*Matrix[k]
            if min_value ==0 or min_value > result:
                min_value = result
        cache[j][k] = min_value
print(cache[1][N])
```

```python
#1149 -DP 두번째 유형.
n = int(input())
p = []

for i in range(n):
    p.append(list(map(int,input().split())))
for i in range(1,len(p)):
    p[i][0] = min(p[i-1][1],p[i-1][2]) + p[i][0]
    p[i][1] = min(p[i-1][0],p[i-1][2]) + p[i][1]
    p[i][2] = min(p[i-1][0],p[i-1][1]) + p[i][2]  

print(min(p[n-1][0],p[n-1][1],p[n-1][2]))
```

```python
#2156 포도주
#첫번째로 해당 순서의 포도주를 마시는 경우, 그 전 포도주를 안먹는 경우와 먹는경우가 있다.
#그리고 해당 순서의 포도주를 마시지 않는 경우가 있다.
n = int(input())
w = [0]
for i in range(n):
    w.append(int(input()))
dp = [0]
dp.append(w[1])
if n > 1:
    dp.append(w[1] + w[2])
for i in range(3, n + 1):
    dp.append(max(dp[i - 1], dp[i - 3] + w[i - 1] + w[i], dp[i - 2] + w[i]))
print(dp[n])
```
