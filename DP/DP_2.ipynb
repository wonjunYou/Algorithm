```python
#1904 - DP 첫 번째 유형
n = int(input())
tiles = [0]*1000000

def Main(n):
    tiles[0] = 1
    tiles[1] = 2
    for i in range(2,n):
        tiles[i] = (tiles[i-1] + tiles[i-2])%15743
    return tiles[n-1]
if(n<=1000000):
    print(Main(n))
```

```python
#1149 -DP 두번째 유형.
n = int(input())
p = []

for i in range(n):
    p.append(list(map(int,input().split())))
for i in range(1,len(p)):
    p[i][0] = min(p[i-1][1],p[i-1][2]) + p[i][0]
    p[i][1] = min(p[i-1][0],p[i-1][2]) + p[i][1]
    p[i][2] = min(p[i-1][0],p[i-1][1]) + p[i][2]  

print(min(p[n-1][0],p[n-1][1],p[n-1][2]))
```

###Dynamic programming    
DP는 크게 4가지 유형으로 분류된다.<br>    
<b>1. Fibonacci를 응용한 유형</b>
    - example: 타일 배치 문제(가로-세로)
    - 해보면 나온다. 누적합. 1+2 ->3 ... (N-2) + (N-1) -> N
    
<b>2. 특수한 조건을 고려해야 하는 경우</b>
    - exmaple: 행렬 경로 문제, 돌 놓기 문제 등 
    - 해결법은 의외로 단순하다. 
    - 특정 index에서 벌어질 수 있는 모든 경우의 수를 생각하고, 
      그에 min, max값을 이용하는 대표적인 유형이다.
      
<b>3. LCS 문제 유형 (LCS, LIS) 등등 -> LCS의 풀이법을 이용</b>
    - BUT, 실생활 문제에 적용시켰을 때, LCS 문제인지 아닌지를 파악하는것이 중요한 관건이다.
    
<b>4. 평범한 배낭(baekjoon-12865), graph의 플로이드-워샬 알고리즘과 같은 특수 유형</b>
    - 모든 경우의 수를 고려하나, 계산에는 순서에 구애받지 않는다.
    - 최종 산출되는 결과만이 유의미하며, 중간 과정에서 구해지는 값들은 의미가 없는 값들이다.
