```python
'''
도미노는 1 x 2 모양의 사각형 하나를 뜻한다. 
가로의 길이 n, 세로의 길이 2인 사각형을 도미노로 채우는 방법의 수는 총 몇 가지인지 계산하는 
알고리즘을 동적 프로그래밍으로 작성하시오. 
최적 부분 구조를 먼저 찾아내고, 이를 실현하는 알고리즘을 작성하시오.   
'''

solutionList = [0]*n
def getSolutions(n):
    solutionList[0] = 1
    solutionList[1] = 2
    for i in range(2,n):
        solutionList[i] = solutionList[i-1] + solutionList[i-2]
    return solutionList[n-1]
#가로의 길이 = n
n = int(input())
print(getSolutions(n))
```

```python
'''
어떤 학생이 n개 문항으로 구성된 시험을 치른다 하자. 
각 문항은 모두 배점이 달라, i문항의 배점은 pi 로 정의되어 있다 하자. 
학생은 반드시 시험 문제를 순서대로(즉, 1,2,3,...,n 의 순서) 풀어야 하며, 
임의의 문제는 skip할 수도 있다(안 풀고 지나갈 수도 있다). 
어떤 문항 i는 너무 어렵기 때문에 
이 문제를 풀고 나서는 힘이 들어 이후 fi개 문항들을 풀 수 없다 하자. 
n개의 문항에 대해 pi, fi 값들이 정의되었다할 때, 
최고로 받을 수 있는 시험 점수를 결정하는 알고리즘을 고안하시오
'''
p = [20,10,5,10]
f= [3,0,0,0]

score = [0]*(len(p)+1)
n = len(p)

for i in range(n-1,-1,-1):
    tmp = p[i]
    
    if(f[i]+i+1 < n):
        tmp = tmp + score[f[i] + i + 1]
    #jump면 어차피 안 풀었을 것.
    jump = score[i+1]
    
    score[i] = max(tmp,jump)
print(score[0])
```

```python
'''
n개의 수로 구성된 수열에서 가장 긴 단조 증가 부분 수열의 길이를 
동적 프로그래밍으로 찾는 알고리즘을 고안하시오. 
예를 들어, 수열 <1, 10, 5, 20, 15, 10, 30, 49, 25, 35, 66, 75, 44>에서 
<1, 10, 10, 30, 49, 66, 75>는 최장 단조 증가 부분 수열 중 하나로 길이가 7이다.

'''

def sortArr(arr):
    mid = n//2
    low_arr = sortArr(arr[:mid])
    high_arr = sortArr(arr[mid:])
    
    sorted_Arr = []
    l=h=0
    while(l<len(low_arr) and h<len(high_arr)):
        if(low_arr[l]<high_arr[h]):
            sorted_arr.append(low_arr[l])
            l+=1
        else:
            sorted_arr.append(high_arr[h])
            h+=1
    sorted_Arr += low_arr[l:]
    sorted_Arr += high_arr[h:]
    
    return sorted_Arr

def LCS(sorted_Arr,arr):
    
    result_Arr = []
    
    for i in range(0,len(sorted_Arr)):
        if(sorted_Arr[i] <= arr[i] and result_Arr[len(result_Arr)-1] <= sorted_Arr[i] ):
            result_Arr.append(arr[i])
    return result_Arr


#cache = [[0]*(n+1) for _ in range(n+1)]

arr = [1,10,5,20,15,10,30,49,25,35,66,75,44]
sorted_Arr = []
n = len(arr)
sorted_Arr = sortArr(arr)
result = LCS(sorted_Arr,arr)

for i in result:
    print(i, end="")

```

```python
def sortArr(arr):
    
    if(len(arr)<2):
        return arr
    
    mid = len(arr)//2
    low_arr = sortArr(arr[:mid])
    high_arr = sortArr(arr[mid:])
    
    sorted_Arr = []
    l=h=0
    while(l<len(low_arr) and h<len(high_arr)):
        if(low_arr[l]<high_arr[h]):
            sorted_Arr.append(low_arr[l])
            l+=1
        else:
            sorted_Arr.append(high_arr[h])
            h+=1
    sorted_Arr += low_arr[l:]
    sorted_Arr += high_arr[h:]
    
    return sorted_Arr

def LCS(sorted_Arr,arr):
    
    result_Arr = []
    
    for i in range(len(sorted_Arr)):
        if(sorted_Arr[i] <= arr[i] ):
            if len(result_Arr) == 0:
                result_Arr.append(arr[i])
            else:
                if result_Arr[len(result_Arr)-1] <= sorted_Arr[i]:
                    result_Arr.append(arr[i])
    return result_Arr


#cache = [[0]*(n+1) for _ in range(n+1)]

arr = [1,10,5,20,15,10,30,49,25,35,66,75,44]
n = len(arr)
sorted_Arr = sortArr(arr)
result = LCS(sorted_Arr,arr)

print(len(result))
```
