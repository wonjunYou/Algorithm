```python
#https://leetcode.com/problems/swap-nodes-in-pairs/

#24. Swap Nodes in Pairs

#Sol 1. using iterative structure

#Runtime: 28 ms
#Memory Usage: 14.1 MB

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def swapPairs(self, head: ListNode) -> ListNode:
        root = prev = ListNode(0)
        prev.next = head
        
        #node swap
        while head and head.next:
            
            b= head.next
            head.next = b.next
            b.next = head
            
            #prev points b
            prev.next = b
            
            #move next for next comparing
            prev = prev.next.next
            head = head.next
            
        return root.next
```

```python
#https://leetcode.com/problems/swap-nodes-in-pairs/

#24. Swap Nodes in Pairs

#Sol 2. using recursive structure

#Runtime: 32 ms
#Memory Usage: 14.1 MB

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def swapPairs(self, head: ListNode) -> ListNode:
        if head and head.next:
          p = head.next

          head.next = self.swapPairs(p.next)
          p.next = head
          return p

        return head
```
