```python
#https://leetcode.com/problems/merge-two-sorted-lists/submissions/

#21. Merge Two Sorted Lists

#Sol 1. using iterative structure
#Runtime: 32 ms
#Memory Usage: 14.2 MB

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:

        #make dummy linked list
        head = ListNode(-1)
        node = head
        
        while l1 != None and l2 != None:
            if l1.val >= l2.val:
                node.next = l2
                l2 = l2.next
            else:
                node.next = l1
                l1 = l1.next
            node = node.next

        #handle last Node
        if l1 is None:
            node.next = l2
        else:
            node.next = l1
            
        return head.next
```

```python
#https://leetcode.com/problems/merge-two-sorted-lists/submissions/

#21. Merge Two Sorted Lists

#Sol 2. using recursive structure
#Runtime: 36 ms
#Memory Usage: 14.3 MB

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
      if not l1 or (l2 and l1.val > l2.val):
        l1, l2 = l2, l1

      if l1:
        l1.next = self.mergeTwoLists(l1.next, l2)

      return l1
```
